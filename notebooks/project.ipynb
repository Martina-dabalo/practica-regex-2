# Práctica de regex

```python
import re
filename = open('../data/baby2004.html')
text = filename.read()
```

## Tarea 1: Encontrar el Año

```python
#line = '<h3 align="center">Popularity in (\d\d\d\d)</h3>'
result = re.search(r'Popularity in (\d\d\d\d)', text)
if result:
    age = result.group(1)
    print(f'Año: {age}')
else:
    age = 'Año no encontrado'
```

## Tarea 2: Extraer Nombres y Ranking

```python
pattern = r'<td>(\d*)</td><td>(\w*)</td><td>(\w*)</td>'
data = []
match = re.findall(pattern, text)

for ranking, boy, girl in match:
    data.append((boy, ranking))
    data.append((girl, ranking))

data[:10], len(data)
```

## Tarea 3: Guardar en un Diccionario

```python
dictionary = {}
for name, ranking in data:
    dictionary[name] = ranking

dictionary
```

## Tarea 4: Lista de Nombres Ordenada

```python
names = list(dictionary.keys())
names.sort()

names
```

## Tarea 5: Formatear la Salida Final

```python
import re
filename = open('../data/baby2004.html')
text = filename.read()

result = re.search(r'Popularity in (\d\d\d\d)', text)
if result:
    age = result.group(1)

pattern = r'<td>(\d*)</td><td>(\w*)</td><td>(\w*)</td>'
data = []
match = re.findall(pattern, text)

for ranking, boy, girl in match:
    data.append((boy, ranking))
    data.append((girl, ranking))

dictionary = {}
for name, ranking in data:
    dictionary[name] = ranking

names = list(dictionary.keys())
names.sort()

name_list = [f'{name} - {ranking}' for name, ranking in data]
name_list.sort()

print(f'{age}, {name_list}')
```
