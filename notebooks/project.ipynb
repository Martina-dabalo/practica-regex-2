# Práctica de regex

```python
import re
filename = open('../data/baby2004.html')
text = filename.read()
```

# Tarea 1

```python

#line = '<h3 align="center">Popularity in (\d\d\d\d)</h3>'
result = re.search(r'Popularity in (\d\d\d\d)', text)
if result:
    age = result.group(1)
    print(f'Año: {age}')
else:
    age = 'Año no encontrado'
```

```python
#line = '<h3 align="center">Popularity in (\d\d\d\d)</h3>'
result = re.search(r'Popularity in (\d\d\d\d)', text)
if result:
    age = result.group(1)
    print(f'Año: {age}')

pattern = r'<td>(\d*)</td><td>(\w*)</td><td>(\w*)</td>'
data = []
match = re.findall(pattern, text)

for ranking, boy, girl in match:
    data.append((boy, ranking))
    data.append((girl, ranking))

data[:10], len(data)
```

```python
dictionary = {}
for name, ranking in data:
    dictionary[name] = ranking

dictionary
```

```python
names = list(dictionary.keys())
names.sort()

names
```

```python
import re
filename = open('../data/baby2008.html')
text = filename.read()

line = '<h3 align="center">Popularity in (\d\d\d\d)</h3>'
result = re.search(r'Popularity in (\d\d\d\d)', text)
if result:
    age = result.group(1)

pattern = r'<td>(\d*)</td><td>(\w*)</td><td>(\w*)</td>'
data = []
match = re.findall(pattern, text)

for ranking, boy, girl in match:
    data.append((boy, ranking))
    data.append((girl, ranking))

dictionary = {}
for name, ranking in data:
    dictionary[name] = ranking

names = list(dictionary.keys())
names.sort()

name_list = [f'{name} - {ranking}' for name, ranking in data]
name_list.sort()

print(f'{age}, {name_list}')
```
